# 2. Weekly graph question

Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
df = pd.DataFrame({'ser1': time_series, 'ser2': time_series_2}, index = datearray)
df.plot.area()
```

One disadvantage of the area plot is that some of the true changes in phosphate might be hidden by the blue of the nitrate. To remedy this I would change the transparency or harmony of the colors so that they could work better together visually. Another issue might be that one would interpet these two as the total chemical components in the water. We could add more labels or titles help clarify some of these gaps. I have a saltwater fish tank that requires careful monitoring of these chemicals, and I know that high levels of nitrate can increase the overall phosphate level as well.

```python
months    = ["January","February","March","April","May","June",
             "July","August","September","October","November","December"]
received  = [160, 180, 240, 150, 180, 160, 130, 200, 160, 140, 150, 180]
processed = [155, 185, 235, 150, 175, 150, 125, 150, 120, 100, 120, 140]

plt.figure(figsize=(10,5))
plt.plot(months, received,  marker="D", markersize=8,
         linewidth=2, color="tab:blue",
         label="Ticket Volume Received")
plt.plot(months, processed, marker="s", markersize=8,
         linewidth=2, color="tab:red",
         label="Ticket Volume Processed")


plt.title("Chart from Page 96")
plt.xlabel("Month")
plt.ylabel("Volume")
plt.ylim(0, 300)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.legend(loc="upper left")


plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```
