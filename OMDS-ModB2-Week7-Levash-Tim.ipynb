```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

concerns = [
    "Engine power low",
    "Tire noise",
    "Engine noise",
    "Seat material",
    "Wind noise",
    "Shift hesitation",
    "Bluetooth sound",
    "Steering play",
    "Bluetooth usability",
    "Audio controls"
]
values = [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]


fig, ax = plt.subplots(figsize=(8, 5))
y_pos = range(len(concerns))[::-1]  
ax.barh(y_pos, values, color="dimgray", height=0.6)


for i, v in enumerate(values[::-1]):
    ax.text(v + 0.2, i, f"{v:.1f}", va="center", ha="left", fontsize=10)


ax.set_yticks(y_pos)
ax.set_yticklabels(concerns[::-1], fontsize=11)
ax.set_xlabel("Concerns per 1,000", fontsize=12)
ax.set_title("Chart from Page 109", fontsize=14, pad=15)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```
